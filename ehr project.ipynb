```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['drugnotetype'].unique()
```

```python
df['drugnotetype'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot('drugnotetype', data = df, palette = 'hls')
plt.show()
```

```python
label_data = df['drugnotetype'].value_counts()

explode = (0.0, 0.0, 0.0, 0.0, 0.15, 0.4)
plt.figure(figsize=(30, 20))
patches, texts, pcts = plt.pie(label_data,
                               labels = label_data.index,
                               colors = ['blue', 'red', 'green', 'orange',
                                        'yellow', 'purple', 'grey'],
                               pctdistance = 0.65,
                               shadow = False,
                               startangle = 0,
                               explode = explode,
                               autopct = '%1.1f%%',
                               textprops={ 'fontsize': 25,
                                           'color': 'black',
                                           'weight': 'bold',
                                           'family': 'serif' })
plt.setp(pcts, color='black')

hfont = {'fontname':'serif', 'weight': 'bold'}
plt.title('Drug Note Type', size=20, **hfont)

centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```
